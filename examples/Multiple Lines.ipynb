This notebook shows how to create two or more independent lines in the same simulation.

```python
from siamese import (
    Source,
    Machine,
    Buffer,
    Line
)

from siamese.distributions import (
    Exponential,
    Triangular
)
```

### Line 1

```python
line1_source = Source(
    name = 'line1_source',
    processing_time = Exponential(1),
    output_buffer = 'line1_buffer1'
)

line1_buffer1 = Buffer(
    name = 'line1_buffer1',
    capacity = 10
)

line1_machine1 = Machine(
    name = 'line1_machine1',
    processing_time = Triangular(min=0.5, mode=1, max=1.5),
    input_buffer = 'line1_buffer1',
    output_buffer = 'line1_sink'
)

line1_sink = Buffer(
    name = 'line1_sink',
    capacity = 100
)
```

### Line 2

```python
line2_source = Source(
    name = 'line2_source',
    processing_time = Exponential(1),
    output_buffer = 'line2_buffer1'
)

line2_buffer1 = Buffer(
    name = 'line2_buffer1',
    capacity = 10
)

line2_machine1 = Machine(
    name = 'line2_machine1',
    processing_time = Triangular(min=0.5, mode=1, max=1.5),
    input_buffer = 'line2_buffer1',
    output_buffer = 'line2_buffer2'
)

line2_buffer2 = Buffer(
    name = 'line2_buffer2',
    capacity = 10
)

line2_machine2 = Machine(
    name = 'line2_machine2',
    processing_time = Triangular(min=0.5, mode=1, max=1.5),
    input_buffer = 'line2_buffer2',
    output_buffer = 'line2_sink'
)

line2_sink = Buffer(
    name = 'line2_sink',
    capacity = 100
)
```

### Lines Wrapper

```python
lines = Line(

    # Line 1 objects
    line1_source,
    line1_buffer1,
    line1_machine1,
    line1_sink,

    # Line 2 objects
    line2_source,
    line2_buffer1,
    line2_machine1,
    line2_buffer2,
    line2_machine2,
    line2_sink

)

lines.plot(seed=15)
```

```python
lines.simulate(100)
lines.report
```
